```python
import geopandas as gpd
from pathlib import Path

# 1. Determine the relative path to the shapefile
# From the 'model' folder, go up one level ("..") and then into the "osm" folder.
shapefile_path = Path("..") / "osm" / "roads.shp"

print(f"Loading shapefile from: {shapefile_path.resolve()}")

# 2. Load the shapefile
road_data = gpd.read_file(shapefile_path)

# 3. Create a clean column without hidden spaces (so 'N 2' becomes 'N2')
road_data['ref_clean'] = road_data['ref'].str.replace(' ', '', regex=False)

# 4. Filter the dataset for the N1 and N2 highways
n1_roads = road_data[road_data['ref_clean'] == 'N1']
n2_roads = road_data[road_data['ref_clean'] == 'N2']

# 5. Calculate the exact geographical intersection
if not n1_roads.empty and not n2_roads.empty:
    # We use union_all() which is the standard for newer GeoPandas versions
    n1_line = n1_roads.geometry.union_all()
    n2_line = n2_roads.geometry.union_all()

    intersection = n1_line.intersection(n2_line)

    print("\n--- Result ---")
    print("The exact geographical intersection between N1 and N2 is:")
    print(intersection)
else:
    print("\nError: N1 or N2 could not be found in the dataset.")
```
