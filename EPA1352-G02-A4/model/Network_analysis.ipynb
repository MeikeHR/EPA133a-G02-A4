```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Lees de S0 dataset in (zorg dat het pad klopt naar jouw map!)
df = pd.read_csv('results simulation/bridges_S0.csv')

# 2. Omdat we 10 replicaties (runs) hebben, nemen we het gemiddelde aantal vrachtwagens
bridge_agg = df.groupby(['bridge_id', 'bridge_name', 'road', 'condition', 'vulnerability_score', 'vuln_level'])['trucks_passed'].mean().reset_index()
bridge_agg.rename(columns={'trucks_passed': 'avg_trucks_passed'}, inplace=True)

# 3. Filter de Top 10 Bottlenecks! (Hoogste vulnerability, daarna meeste vrachtwagens)
top_bottlenecks = bridge_agg.sort_values(by=['vulnerability_score', 'avg_trucks_passed'], ascending=[False, False]).head(10)

print("🏆 TOP 10 MEEST KRITIEKE EN KWETSBARE BRUGGEN:")
display(top_bottlenecks[['bridge_id', 'bridge_name', 'road', 'condition', 'vuln_level', 'avg_trucks_passed']])

# 4. Maak de Visualisatie voor in het verslag
plt.figure(figsize=(12, 7))

# Plot alle bruggen
sns.scatterplot(
    data=bridge_agg,
    x='avg_trucks_passed',
    y='vulnerability_score',
    hue='vuln_level',
    palette={'Low': 'green', 'Medium-Low': 'gold', 'Medium-High': 'darkorange', 'High': 'crimson'},
    alpha=0.6,
    s=60
)

# Highlight de Top 10 met een zwarte rand
plt.scatter(
    top_bottlenecks['avg_trucks_passed'],
    top_bottlenecks['vulnerability_score'],
    color='none',
    edgecolor='black',
    linewidth=2,
    s=150,
    label='Top 10 Bottlenecks'
)

plt.title('Criticality vs Vulnerability of Bridges in Bangladesh (Simulated Output)', fontsize=14, fontweight='bold')
plt.xlabel('Simulated Criticality (Average Trucks Passed)', fontsize=12)
plt.ylabel('Physical Vulnerability (Vulnerability Score)', fontsize=12)
plt.legend(title='Vulnerability Level')
plt.grid(True, linestyle='--', alpha=0.5)

# Sla hem netjes op voor in je PDF
plt.tight_layout()
plt.savefig('scatter_bottlenecks.png', dpi=300)
plt.show()
```
